```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect(r'/Users/danielwanganga/Documents/GitHub/PROJECT_FRONTLINE/INTERN_PROJECT/leave_management.db')
```

```python
employee_data = pd.read_csv(r"/Users/danielwanganga/Documents/GitHub/PROJECT_FRONTLINE/INTERN_PROJECT/data/employee_table.csv")
leave_entry_data = pd.read_csv(r"/Users/danielwanganga/Documents/GitHub/PROJECT_FRONTLINE/INTERN_PROJECT/data/leave_entries.csv")
leave_entitlements_data = pd.read_csv(r"/Users/danielwanganga/Documents/GitHub/PROJECT_FRONTLINE/INTERN_PROJECT/data/leave_entitlements_data.csv")

```

```python
employee_data.to_sql(name="employee_table",con=conn,if_exists='replace',index=False)
leave_entry_data.to_sql(name="leave_entry",con=conn,if_exists='replace',index=False)
leave_entitlements_data.to_sql(name="leave_entitlements_data",con=conn,if_exists='replace',index=False)
```
